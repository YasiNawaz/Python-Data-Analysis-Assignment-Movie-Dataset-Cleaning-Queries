### **Question 21**
(Visualization) IMDB_Rating ka distribution dekhne ke liye ek neat Histogram plot karein.

### **Question 22**

```python
top_directors = movie['Director'].value_counts().head(10)

plt.figure(figsize=(10, 6))

plt.bar(top_directors.index, top_directors.values)

plt.title('Top 10 Directors by Total Movies')
plt.xlabel('Directors')
plt.ylabel('Total Movies')

plt.xticks(rotation=45)

plt.show()
```
